```python
import openai
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, set_seed
```

```python
import torch
torch.set_default_tensor_type(torch.cuda.FloatTensor)

model = AutoModelForCausalLM.from_pretrained('bigscience/bloom-1b7', use_cache=True)
tokenizer = AutoTokenizer.from_pretrained('bigscience/bloom-1b7')
```

```python
set_seed(424242)
```

```python
openai.api_key = #(api-key generated from openai)
```

```python
# Auto complete job prompt using bloom
prompt = 'Electrical engineer with experience at Tesla  '
inputids = tokenizer(prompt, return_tensors='pt').to(0)

sample = model.generate(**inputids, max_length=200, num_beams = 2, num_beam_groups = 2, top_k = 1, temperature = 0.9, repetition_penalty = 3.5, diversity_penalty = 0.9)

prompt1 = tokenizer.decode(sample[0], truncate_before_pattern=[r"\n\n^#$", "^'''", "\n\n\n"])

print(prompt1)
```

```python
# Generate resume response using GPT3 
response = openai.Completion.create(model="text-davinci-003", prompt= 'Build a one page professional resume from the following summary: ' + prompt1 + 'Resume: \n', 
                                    max_tokens=2500, top_p = 1, best_of=5, presence_penalty=1.5)
print(response.choices[0].text)
```
